```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
sentences_df = pd.read_csv('.\dataset\csv\sentencing.csv', delimiter=';')
sentences_df.head()


# Function to categorize offenses based on string matching
def categorize_offense(offense):
    offense = str(offense).lower()  # Convert to lowercase for case-insensitive matching
    
    # Categories based on keywords
    if re.search(r'\bkidnapping\b', offense):
        return 'Kidnapping'
    elif re.search(r'\bmurder\b|\bassault\b', offense):
        return 'Murder/Assault'
    elif re.search(r'\brape\b|\bsex aslt\b', offense):
        return 'Sex Offenses'
    elif re.search(r'\btheft\b', offense):
        return 'Theft'
    elif re.search(r'\bdrug\b|\bmfg\b|\bdel\b', offense):
        return 'Drug Offenses'
    elif re.search(r'\binvasion\b', offense):
        return 'Invasion'
    elif re.search(r'\bweapon\b', offense):
        return 'Weapon'
    elif re.search(r'\bfraud\b', offense):
        return 'Fraud'
    elif re.search(r'\bbattery\b', offense):
        return 'Battery'
    elif re.search(r'\bburglary\b', offense):
        return 'Burglary'
    elif re.search(r'\brobbery\b', offense):
        return 'Robbery'

# Apply the categorization function to the 'offense' column
sentences_df['offense_category'] = sentences_df['offense'].apply(categorize_offense)


# Count the number of inmates per offense category
offense_counts = sentences_df['offense_category'].value_counts()


# Plot the number of inmates for each offense category
offense_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Inmates for Each Offense Category')
plt.xlabel('Offense Category')
plt.ylabel('Number of Inmates')
plt.xticks(rotation=45, ha='right')  # Rotate labels if they overlap
plt.tight_layout()
plt.show()
```
